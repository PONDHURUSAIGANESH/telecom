```python
var = list(df.select_dtypes(exclude = ['object', 'bool']).columns)
unvar = ['Churn', 'International plan', 'Voice mail plan', 'Area Code']
var = [r for r in var if r not in unvar]
var
plt.figure(figsize = (20,170))
l_row = 50
l_pos = 1
for col in nuvar:
    plt.subplot(l_row, 4, l_pos)
    sns.boxplot(y = col, data = df)
    l_pos = l_pos + 1
plt.tight_layout()
plt.show()
```

```python
for col in var:
    Q1 = df[col].quantile(0.05)
    Q3 = df[col].quantile(0.95)
    df[col][df[col] <= Q1] = Q1
    df[col][df[col] >= Q3] = Q3
```

```python
# Creating function for Bivariate Analysis
def bivariate(lvar1, lvar2):
    plt.figure(figsize = (15, 5))
    plt.subplot(1, 2, 1)
    plt.suptitle('Bivariate Analysis of Variables ' + lvar1 + ' And ' + lvar2, fontsize = 15)
    sns.boxplot(lvar1, lvar2, data = df)
    plt.subplot(1, 2, 2)
    sns.scatterplot(lvar1, lvar2, data = df)
    plt.tight_layout()
    plt.show()
```

```python
bivariate('Churn', 'Total day charge')
```

```python
bivariate('Churn', 'Account length')
```
